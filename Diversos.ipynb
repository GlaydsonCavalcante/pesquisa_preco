```python
# CORRIGE O ERRO DE CLASSIFICA√á√ÉO DE MODELOS NO BANCO DE DADOS
import sqlite3
import pandas as pd
import os
import json
import time
import random
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from google import genai
from google.genai import types
from dotenv import load_dotenv

# --- 1. CONFIGURA√á√ÉO ---
DB_PATH = os.path.join("data", "historico_precos.db")
CSV_PATH = os.path.join("data", "modelos_alvo.csv")
# Se quiser ver o navegador rodando, deixe HEADLESS = False
HEADLESS = False 

load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

if not API_KEY:
    load_dotenv("../.env")
    API_KEY = os.getenv("GEMINI_API_KEY")

if API_KEY:
    client = genai.Client(api_key=API_KEY)
    print("‚úÖ Cliente Gemini Conectado!")
else:
    print("‚ùå ERRO: API Key n√£o encontrada.")

# --- 2. GEST√ÉO DO BANCO DE DADOS (MIGRATION) ---
def garantir_coluna_verificado():
    """Cria a coluna 'verificado' se ela n√£o existir."""
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    try:
        # Tenta selecionar a coluna. Se falhar, √© porque n√£o existe.
        cursor.execute("SELECT verificado FROM precos LIMIT 1")
    except sqlite3.OperationalError:
        print("üõ†Ô∏è Criando coluna 'verificado' no banco de dados...")
        cursor.execute("ALTER TABLE precos ADD COLUMN verificado INTEGER DEFAULT 0")
        conn.commit()
    finally:
        conn.close()

garantir_coluna_verificado()

# --- 3. CONFIGURA√á√ÉO DO SELENIUM (O "ROB√î VISUAL") ---
def iniciar_driver():
    chrome_options = Options()
    if HEADLESS:
        chrome_options.add_argument("--headless")
    
    # Truques para evitar detec√ß√£o b√°sica
    chrome_options.add_argument("--disable-blink-features=AutomationControlled") 
    chrome_options.add_argument("--start-maximized")
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

# --- 4. EXTRATOR COM SELENIUM (Evolu√ß√£o do Scrape) ---
def extrair_dados_com_selenium(driver, url):
    dados = {"titulo": "", "descricao": "", "sucesso": False}
    
    try:
        driver.get(url)
        # Espera aleat√≥ria humana
        time.sleep(random.uniform(2, 4))
        
        # 1. Tenta pegar o T√≠tulo (Espera at√© aparecer)
        try:
            h1 = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "h1"))
            )
            dados["titulo"] = h1.text.strip()
        except:
            pass # T√≠tulo vazio

        # 2. Tenta pegar a Descri√ß√£o (Mercado Livre usa classes espec√≠ficas)
        try:
            # Tenta seletor padr√£o do ML
            desc_element = driver.find_element(By.CLASS_NAME, "ui-pdp-description__content")
            dados["descricao"] = desc_element.text.strip()[:3000]
        except:
            try:
                # Tentativa secund√°ria (alguns layouts mudam)
                desc_element = driver.find_element(By.CLASS_NAME, "ui-pdp-description")
                dados["descricao"] = desc_element.text.strip()[:3000]
            except:
                dados["descricao"] = "Descri√ß√£o n√£o carregou ou vazia."

        # Valida√ß√£o b√°sica: Se tiver t√≠tulo, consideramos sucesso na leitura
        if dados["titulo"]:
            dados["sucesso"] = True
            
    except Exception as e:
        print(f"‚ö†Ô∏è Erro no navegador: {e}")
        
    return dados

# --- 5. INTELIG√äNCIA ARTIFICIAL ---
def carregar_whitelist():
    try:
        df = pd.read_csv(CSV_PATH, on_bad_lines='skip')
        return [str(m).strip() for m in df['modelo'].dropna().unique().tolist()]
    except:
        return []

MODELOS_VALIDOS = carregar_whitelist()

def auditar_com_ia(dados_site, preco, modelo_db):
    prompt = f"""
    Atue como Auditor de Dados de Instrumentos Musicais.
    
    DADOS DO AN√öNCIO (Rec√©m coletados):
    T√≠tulo: "{dados_site['titulo']}"
    Pre√ßo: R$ {preco}
    Descri√ß√£o: "{dados_site['descricao'][:2000]}..."
    Modelo Cadastrado (Pode estar errado): {modelo_db}
    
    WHITELIST (MODELOS ACEITOS):
    {json.dumps(MODELOS_VALIDOS)}
    
    DEFINI√á√ÉO DE ESTADOS (estado_detalhado):
    - novo: Lacrado, loja oficial.
    - otimo_estado: Usado perfeito.
    - funcional: Marcas de uso normais, 100% operacional.
    - semifuncional: Defeito leve (1 tecla muda, bot√£o falhando, som baixo).
    - nao_funcional: N√£o liga, quebrado.

    REGRAS DE DECIS√ÉO:
    1. Acess√≥rios/Pe√ßas -> acao: "DESATIVAR".
    2. "Kits" (ex: "Kit Piano P45 + Suporte") -> √â o PIANO (esta_lista: true).
    3. esta_lista: Responda 'true' SE E SOMENTE SE o modelo real estiver na Whitelist.
    4. modelo_real: Identifique o nome real, corrigindo se necess√°rio.
    
    SA√çDA JSON OBRIGAT√ìRIA:
    {{
        "acao": "CORRIGIR" | "DESATIVAR" | "MANTER",
        "modelo_real": "Nome real do produto",
        "estado": "novo" | "otimo_estado" | "funcional" | "semifuncional" | "nao_funcional",
        "reparo": float (0 se ok, valor estimado se defeito),
        "esta_lista": boolean,
        "justificativa": "Explica√ß√£o curta"
    }}
    """

    try:
        response = client.models.generate_content(
            model='gemini-2.5-flash', 
            contents=prompt,
            config=types.GenerateContentConfig(temperature=0.0, response_mime_type='application/json')
        )
        return json.loads(response.text)
    except:
        return None

# --- 6. MOTOR DE EXECU√á√ÉO (SEQUENCIAL PARA VOC√ä ACOMPANHAR) ---
def executar_auditoria_visual():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    
    # Pega itens que ainda N√ÉO foram verificados (verificado=0) OU que est√£o ativos
    # Se quiser for√ßar re-verifica√ß√£o de tudo, remova o "AND verificado=0"
    print("üîç Buscando itens pendentes...")
    cursor.execute("SELECT id, modelo, link, preco FROM precos WHERE ativo = 1 ORDER BY id DESC")
    fila = cursor.fetchall()
    
    print(f"üöÄ Iniciando Auditoria Visual em {len(fila)} itens.")
    print("üëÄ O navegador vai abrir. N√£o o feche!")
    
    driver = iniciar_driver()
    
    processed = 0
    updated = 0
    
    try:
        for item in fila:
            id_db, modelo_db, link_db, preco_db = item
            processed += 1
            
            print(f"\n[{processed}/{len(fila)}] Acessando ID {id_db}...")
            
            # 1. Scrape Visual
            dados = extrair_dados_com_selenium(driver, link_db)
            
            # Se falhou em ler o site (ex: p√°gina n√£o existe), marca n√£o verificado e pula
            if not dados["sucesso"]:
                print(f"   ‚ùå Falha ao ler site. Marcando como N√ÉO verificado.")
                cursor.execute("UPDATE precos SET verificado = 0 WHERE id = ?", (id_db,))
                conn.commit()
                continue
                
            print(f"   ‚úÖ Site lido: {dados['titulo'][:40]}...")
            
            # 2. IA
            res = auditar_com_ia(dados, preco_db, modelo_db)
            if not res:
                print("   ‚ö†Ô∏è Falha na IA. Pulando.")
                continue
                
            # 3. Processa Decis√£o
            acao = res.get('acao', 'MANTER')
            novo_modelo = res.get('modelo_real', modelo_db)
            novo_estado = res.get('estado', 'funcional')
            novo_reparo = res.get('reparo', 0)
            esta_lista = res.get('esta_lista', False)
            justif = res.get('justif', '')
            
            nova_condicao = "Novo" if novo_estado == 'novo' else "Usado"
            
            # Regra de Ouro: Desativar se a IA mandar OU se n√£o estiver na lista
            novo_ativo = 1
            if acao == 'DESATIVAR' or not esta_lista:
                novo_ativo = 0
            
            # 4. Atualiza DB e marca VERIFICADO = 1
            sql = """
                UPDATE precos 
                SET modelo = ?, condicao = ?, estado_detalhado = ?, custo_reparo = ?, 
                    ai_analise = ?, ativo = ?, verificado = 1
                WHERE id = ?
            """
            cursor.execute(sql, (novo_modelo, nova_condicao, novo_estado, novo_reparo, f"Audit: {justif}", novo_ativo, id_db))
            conn.commit()
            updated += 1
            
            status_icon = "üü¢" if novo_ativo else "üî¥"
            print(f"   {status_icon} Resultado: {acao} | Modelo: {novo_modelo} | {novo_estado}")

    except KeyboardInterrupt:
        print("\nüõë Parando execu√ß√£o a pedido do usu√°rio...")
    finally:
        driver.quit()
        conn.close()
        print(f"\nüèÅ Sess√£o Finalizada. {updated} itens atualizados com sucesso.")

if __name__ == "__main__":
    executar_auditoria_visual()
```
