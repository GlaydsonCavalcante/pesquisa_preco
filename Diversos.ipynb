```python
# ATIVA TODOS OS MODELOS INATIVOS NO BANCO DE DADOS
import sqlite3
import os

# Caminho do banco de dados
DB_PATH = os.path.join("data", "historico_precos.db")

# Verifica se o arquivo existe
if not os.path.exists(DB_PATH):
    print(f"‚ùå Erro: Banco de dados n√£o encontrado em {DB_PATH}")
else:
    try:
        # Conex√£o com o banco
        conn = sqlite3.connect(DB_PATH)
        cursor = conn.cursor()

        # 1. Verificar quantos est√£o inativos antes
        cursor.execute("SELECT COUNT(*) FROM precos WHERE ativo = 0")
        resultado = cursor.fetchone()
        inativos_antes = resultado[0] if resultado else 0
        print(f"üìä Itens inativos encontrados: {inativos_antes}")

        # 2. Executar o UPDATE geral
        print("üîÑ Ativando todos os registros...")
        cursor.execute("UPDATE precos SET ativo = 1")
        conn.commit()

        # 3. Confirma√ß√£o
        cursor.execute("SELECT COUNT(*) FROM precos WHERE ativo = 1")
        resultado_final = cursor.fetchone()
        total_ativos = resultado_final[0] if resultado_final else 0
        print(f"‚úÖ Sucesso! Total de itens ativos agora: {total_ativos}")

        conn.close()
        
    except Exception as e:
        print(f"‚ùå Erro ao manipular o banco: {e}")
```
