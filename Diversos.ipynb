```python
# CORRIGE O ERRO DE CLASSIFICA√á√ÉO DE MODELOS NO BANCO DE DADOS
import sqlite3
import pandas as pd
import os
import json
import time
import requests
from bs4 import BeautifulSoup
from datetime import timedelta
from google import genai
from google.genai import types
from dotenv import load_dotenv

# --- 1. CONFIGURA√á√ÉO ---
DB_PATH = os.path.join("data", "historico_precos.db")
CSV_PATH = os.path.join("data", "modelos_alvo.csv")

# Carrega API Key
load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

if not API_KEY:
    load_dotenv("../.env")
    API_KEY = os.getenv("GEMINI_API_KEY")

if API_KEY:
    client = genai.Client(api_key=API_KEY)
    print("‚úÖ Cliente Gemini Conectado!")
else:
    print("‚ùå ERRO: API Key n√£o encontrada.")

# --- 2. CARREGAMENTO DA WHITELIST ---
def carregar_modelos_validos():
    try:
        df = pd.read_csv(CSV_PATH, on_bad_lines='skip')
        modelos = df['modelo'].dropna().unique().tolist()
        return modelos
    except Exception as e:
        print(f"‚ùå Erro CSV: {e}")
        return []

MODELOS_VALIDOS = carregar_modelos_validos()

# --- 3. EXTRATOR DE DADOS (Scraper) ---
def extrair_dados_anuncio(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    dados = {"titulo": "N√£o identificado", "descricao": "Sem descri√ß√£o"}
    try:
        response = requests.get(url, headers=headers, timeout=5)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # T√≠tulo
            h1 = soup.find('h1', class_='ui-pdp-title') or soup.find('h1')
            if h1: dados["titulo"] = h1.get_text().strip()
            
            # Descri√ß√£o
            desc_div = soup.find('p', class_='ui-pdp-description__content') or soup.find('div', class_='ui-pdp-description')
            if desc_div:
                dados["descricao"] = desc_div.get_text(separator="\n").strip()[:3000] # Limite para n√£o estourar token
    except:
        pass
    return dados

# --- 4. CONSULTA √Ä IA (Prompt Ajustado) ---
def auditoria_completa_ia(id_item, modelo_db, link, preco, dados_site):
    prompt = f"""
    Atue como Auditor T√©cnico de Instrumentos Musicais.
    
    AN√öNCIO REAL:
    - T√≠tulo: "{dados_site['titulo']}"
    - Pre√ßo: R$ {preco}
    - Descri√ß√£o: "{dados_site['descricao']}"
    - Modelo Atual no Banco: {modelo_db}
    
    WHITELIST (MODELOS V√ÅLIDOS):
    {json.dumps(MODELOS_VALIDOS)}
    
    TAREFAS:
    1. VALIDA√á√ÉO: 
       - √â acess√≥rio (capa, fonte, suporte)? -> DESATIVAR.
       - Pre√ßo irreal (golpe)? -> DESATIVAR.
       - √â Piano? Qual modelo EXATO da Whitelist? (Se n√£o estiver na lista -> DESATIVAR).
    
    2. ESTADO DETALHADO (Ler descri√ß√£o procurando defeitos):
       - 'novo': Lacrado.
       - 'otimo_estado': Usado perfeito.
       - 'funcional': Marcas est√©ticas, funcionamento 100%.
       - 'semifuncional': Defeito leve (tecla, som, bot√£o).
       - 'nao_funcional': Defeito grave/N√£o liga.
       
    3. CUSTO REPARO:
       - Estime valor (R$) se tiver defeito. Se ok, 0.
       
    4. JUSTIFICATIVA (CRUCIAL):
       - Explique tecnicamente o motivo da classifica√ß√£o em at√© 2 frases.
    
    SA√çDA JSON:
    {{
        "acao": "CORRIGIR" | "DESATIVAR" | "MANTER",
        "modelo_oficial": "Nome da Whitelist ou null",
        "estado_detalhado": "novo" | "otimo_estado" | "funcional" | "semifuncional" | "nao_funcional",
        "custo_reparo": float,
        "justificativa": "Texto explicativo curto"
    }}
    """
    
    for i in range(3): # 3 Tentativas
        try:
            time.sleep(1.5)
            response = client.models.generate_content(
                model='gemini-3-flash-preview',      # gemini-2.5-flash 
                contents=prompt,
                config=types.GenerateContentConfig(temperature=0.0, response_mime_type='application/json')
            )
            return json.loads(response.text)
        except:
            time.sleep(2)
    return None

# --- 5. MOTOR DE EXECU√á√ÉO (Atualiza√ß√£o Completa) ---
def executar_auditoria_profunda():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    
    cursor.execute("SELECT id, modelo, link, preco FROM precos WHERE ativo = 1")
    registros = cursor.fetchall()
    total = len(registros)
    
    print(f"\nüöÄ Iniciando AUDITORIA em {total} registros...")
    print(f"üìù Atualizando colunas: modelo, condicao, estado, reparo, analise.\n")
    
    start_time = time.time()
    updates = 0
    
    for i, (id_db, modelo_db, link_db, preco_db) in enumerate(registros):
        # 1. Dados do Site
        dados_site = extrair_dados_anuncio(link_db)
        
        # Progresso
        elapsed = time.time() - start_time
        restante = (elapsed / (i + 1)) * (total - (i + 1))
        print(f"‚è≥ [{i+1}/{total}] ID {id_db} | {dados_site['titulo'][:30]}... | Falta: {str(timedelta(seconds=int(restante)))}", end="\r")
        
        # 2. IA
        res = auditoria_completa_ia(id_db, modelo_db, link_db, preco_db, dados_site)
        if not res: continue

        acao = res.get('acao', 'MANTER')
        novo_modelo = res.get('modelo_oficial')
        novo_estado = res.get('estado_detalhado', 'funcional')
        novo_reparo = res.get('custo_reparo', 0)
        justificativa = res.get('justificativa', 'Sem justificativa')[:700] # Limite 700 chars
        
        # L√≥gica de Condi√ß√£o (Novo vs Usado)
        nova_condicao = "Novo" if novo_estado == 'novo' else "Usado"
        
        sql = None
        params = None
        log_resumo = ""
        
        # 3. Decis√£o
        if acao == 'DESATIVAR':
            log_resumo = f"üî¥ DESATIVADO | Motivo: {justificativa[:50]}..."
            sql = "UPDATE precos SET ativo = 0, ai_analise = ? WHERE id = ?"
            params = (f"Auditado: {justificativa}", id_db)
            
        elif acao == 'CORRIGIR':
            if novo_modelo and novo_modelo in MODELOS_VALIDOS:
                log_resumo = f"üü¢ CORRIGIDO | {modelo_db} -> {novo_modelo} | {nova_condicao} ({novo_estado})"
                sql = """
                    UPDATE precos 
                    SET modelo = ?, condicao = ?, estado_detalhado = ?, custo_reparo = ?, ai_analise = ? 
                    WHERE id = ?
                """
                params = (novo_modelo, nova_condicao, novo_estado, novo_reparo, f"Auditado: {justificativa}", id_db)
            else:
                log_resumo = f"üü† FORA DA LISTA | Modelo '{novo_modelo}' n√£o monitorado."
                sql = "UPDATE precos SET ativo = 0, ai_analise = ? WHERE id = ?"
                params = (f"Fora da lista: {justificativa}", id_db)
        
        elif acao == 'MANTER':
             log_resumo = f"üîµ REVISADO | {nova_condicao} | {novo_estado} | Reparo: R${novo_reparo}"
             sql = """
                UPDATE precos 
                SET condicao = ?, estado_detalhado = ?, custo_reparo = ?, ai_analise = ? 
                WHERE id = ?
             """
             params = (nova_condicao, novo_estado, novo_reparo, f"Auditado: {justificativa}", id_db)

        # 4. Commit
        if sql:
            print(f"\n   ID {id_db}: {log_resumo}")
            cursor.execute(sql, params)
            conn.commit()
            updates += 1

    print(f"\n\nüèÅ FIM! {updates} registros atualizados.")
    conn.close()

if __name__ == "__main__":
    executar_auditoria_profunda()
```
