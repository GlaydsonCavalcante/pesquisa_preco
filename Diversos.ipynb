```python
# CORRIGE O ERRO DE CLASSIFICA√á√ÉO DE MODELOS NO BANCO DE DADOS
import sqlite3
import pandas as pd
import os
import json
import time
import random
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from google import genai
from google.genai import types
from dotenv import load_dotenv
from concurrent.futures import ThreadPoolExecutor, as_completed

# --- 1. CONFIGURA√á√ÉO ---
DB_PATH = os.path.join("data", "historico_precos.db")
CSV_PATH = os.path.join("data", "modelos_alvo.csv")
MAX_WORKERS = 4 

load_dotenv()
API_KEY = os.getenv("GEMINI_API_KEY")

if API_KEY:
    client = genai.Client(api_key=API_KEY)
    print("‚úÖ Cliente Gemini Conectado!")
else:
    print("‚ùå ERRO: API Key n√£o encontrada.")

# --- 2. EXTRATOR HTTP (Modo Silencioso) ---
def extrair_dados_http(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
        'Referer': 'https://www.google.com/',
    }
    
    dados = {"titulo": "", "descricao": "", "valido": False}
    
    try:
        time.sleep(random.uniform(1, 2))
        response = requests.get(url, headers=headers, timeout=5)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Verifica bloqueio de login
            page_title = soup.title.string.lower() if soup.title else ""
            if "iniciar sess√£o" in page_title or "login" in page_title:
                return dados # Retorna inv√°lido (Bloqueado)
            
            # T√≠tulo
            h1 = soup.find('h1', class_='ui-pdp-title') or soup.find('h1')
            if h1:
                dados["titulo"] = h1.get_text().strip()
                
                # Descri√ß√£o (Muitas vezes n√£o carrega no modo silencioso)
                desc = soup.find('p', class_='ui-pdp-description__content') or soup.find('div', class_='ui-pdp-description')
                
                if desc:
                    dados["descricao"] = desc.get_text(separator="\n").strip()[:1500]
                else:
                    # AQUI EST√Å O SEGREDOS: Avisamos explicitamente que n√£o tem descri√ß√£o
                    dados["descricao"] = "DESCRI√á√ÉO_INDISPONIVEL_HTTP"
                
                dados["valido"] = True

    except Exception:
        pass 
        
    return dados

# --- 3. IA (Prompt Atualizado com 'Indefinido') ---
def carregar_whitelist():
    try:
        df = pd.read_csv(CSV_PATH, on_bad_lines='skip')
        return [str(m).strip() for m in df['modelo'].dropna().unique().tolist()]
    except:
        return []

MODELOS_VALIDOS = carregar_whitelist()

def auditar_com_ia(dados_site, preco, modelo_db):
    prompt = f"""
    Atue como Auditor de Dados.
    
    AN√öNCIO:
    T√≠tulo: "{dados_site['titulo']}"
    Pre√ßo: R$ {preco}
    Descri√ß√£o: "{dados_site['descricao']}"
    Modelo DB: {modelo_db}
    
    WHITELIST: {json.dumps(MODELOS_VALIDOS)}
    
    ESTADOS POSS√çVEIS:
    - novo: Lacrado/Loja.
    - otimo_estado: Usado perfeito (exige descri√ß√£o clara).
    - funcional: Marcas de uso (exige descri√ß√£o clara).
    - semifuncional: Defeito leve.
    - nao_funcional: Defeito grave.
    - indefinido: SE a descri√ß√£o for "DESCRI√á√ÉO_INDISPONIVEL_HTTP" ou vazia, use esta op√ß√£o. N√£o adivinhe.
    
    REGRAS:
    1. Acess√≥rios/Pe√ßas -> DESATIVAR.
    2. Kits Piano -> MANTER (esta_lista: true).
    3. Se n√£o tiver descri√ß√£o, marque 'estado': 'indefinido'.
    4. Identifique o modelo real.
    
    JSON:
    {{
        "acao": "CORRIGIR" | "DESATIVAR" | "MANTER",
        "modelo_real": "Nome real",
        "estado": "novo" | "otimo_estado" | "funcional" | "semifuncional" | "nao_funcional" | "indefinido",
        "reparo": float,
        "esta_lista": boolean,
        "justif": "Texto curto"
    }}
    """
    
    try:
        response = client.models.generate_content(
            model='gemini-2.5-flash', 
            contents=prompt,
            config=types.GenerateContentConfig(temperature=0.0, response_mime_type='application/json')
        )
        return json.loads(response.text)
    except:
        return None

# --- 4. EXECUTOR ---
def processar_item(item):
    id_db, modelo_db, link_db, preco_db = item
    
    dados = extrair_dados_http(link_db)
    
    if not dados["valido"]:
        return None 
    
    res = auditar_com_ia(dados, preco_db, modelo_db)
    if not res: return None
    
    acao = res.get('acao', 'MANTER')
    novo_modelo = res.get('modelo_real', modelo_db)
    novo_estado = res.get('estado', 'indefinido')
    reparo = res.get('reparo', 0)
    esta_lista = res.get('esta_lista', False)
    justif = res.get('justif', '')
    
    # Mapeamento da Condi√ß√£o "Macro"
    if novo_estado == 'novo':
        nova_condicao = "Novo"
    elif novo_estado == 'indefinido':
        nova_condicao = "A Verificar" # Fica claro no Dashboard que precisamos olhar
    else:
        nova_condicao = "Usado"
    
    novo_ativo = 1
    if acao == 'DESATIVAR' or not esta_lista:
        novo_ativo = 0
        
    sql = """
        UPDATE precos 
        SET modelo = ?, condicao = ?, estado_detalhado = ?, custo_reparo = ?, 
            ai_analise = ?, ativo = ?, verificado = 1
        WHERE id = ?
    """
    params = (novo_modelo, nova_condicao, novo_estado, reparo, f"Audit HTTP: {justif}", novo_ativo, id_db)
    
    icon = "üü¢" if novo_ativo else "üî¥"
    if novo_estado == 'indefinido' and novo_ativo: icon = "üü°" # Amarelo se for ativo mas incerto
    
    log = f"{icon} ID {id_db}: {acao} | {novo_modelo} ({novo_estado})"
    return {"sql": sql, "params": params, "log": log}

def executar_sanitizacao_indefinida():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    
    # Garante coluna verificado
    try: cursor.execute("SELECT verificado FROM precos LIMIT 1")
    except: cursor.execute("ALTER TABLE precos ADD COLUMN verificado INTEGER DEFAULT 0")

    cursor.execute("SELECT id, modelo, link, preco FROM precos WHERE ativo = 1 AND verificado = 0")
    lista = cursor.fetchall()
    conn.close()
    
    print(f"üöÄ SANITIZA√á√ÉO HTTP (COM ESTADO 'INDEFINIDO') - {len(lista)} itens")
    
    updates = 0
    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        futures = {executor.submit(processar_item, item): item for item in lista}
        
        for future in as_completed(futures):
            res = future.result()
            if res:
                print(res["log"])
                c_temp = sqlite3.connect(DB_PATH)
                c_temp.execute(res["sql"], res["params"])
                c_temp.commit()
                c_temp.close()
                updates += 1

    print(f"\nüèÅ FIM! {updates} itens processados.")

if __name__ == "__main__":
    executar_sanitizacao_indefinida()
```
